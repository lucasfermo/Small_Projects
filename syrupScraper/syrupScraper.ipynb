```python
def syrupScrapeNew(url,index):
    
    index=str(dateIndex)
    
    res=requests.get(url,allow_redirects=False)
    print("Requesting" + str(res.url))
    soup=bs4.BeautifulSoup(res.text,"html.parser")
    

    #print(soup)
    
    rows=soup.select(".elementor-col-66")
    points=[]
    
    #print(rows)
    
    date=soup.select(".published")
    date=str(date).split(">")[1][:10]
    
    #Iterate through each element
    for row in rows:
        split=str(row).split("<li>")
        new=True
        
        
        for i,s in enumerate(split):

            #Get brand name
            if i==0:
                brand=s.split("<h3>")[1].split("<")[0].replace(":","")

            #Check if ongoing or nw
            if '>Ongoing' in s:
                new=False
            if "href" in s: s=s.split(">")[1]
            if "</li>" in s:
                s=s.split("</")[0].replace("</p>","")
                if new:
                    temp=brand+"~" + date + "~" + index+"~"+ "NEW~" + s
                    points.append(temp)
                else:
                    temp=brand+"~"+date+"~" + index+ "~"+"ONGOING~" +  s
                    points.append(temp)

    return points

def syrupScrapeOld(url,index):
    
    index=str(index)
    
    res=requests.get(url)
    soup=bs4.BeautifulSoup(res.text,"html.parser")
    
    rows=soup.select('.article-content')
    rows=str(rows).replace("h3","h2")
    rows=str(rows).split("<h2")[2:]
   # print(rows)

    date=soup.select(".published")
    date=str(date).split(">")[1][:10]
    
    points=[]
    
    for i in rows[:len(rows)-1]:
        brand=i.split("</")[0].replace(">","").replace(":","")
        b=i.split('<br/>')
        new=True
        for j in b:
            j=j.replace("• ",'').replace("</p>","")
            if "href" in j: 
                j=j.split(">")[1]
            
            
            if 'g>Ongoing' in j:
                new=False
            elif ">New" in j:
                pass
            elif new:
                temp=brand+"~"+date+"~"+index+"~"+"NEW"+"~"+j
                points.append(temp)
                
            else:
                temp=brand+"~"+date+"~"+index+"~"+"ONGOING"+"~"+j
                points.append(temp)
    
    return points
            
        
```
